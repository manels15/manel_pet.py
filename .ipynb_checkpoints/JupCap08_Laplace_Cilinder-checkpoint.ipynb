O objetivo deste notebook é resolver a equação de Laplace em duas dimensões para uma função $u:\mathbb{R}^2\to\mathbb{R}$, onde é especificado o valor de $u(1,\theta)=f(\theta)$. A função $u(r,\theta)$ está especificada em coordenadas polares. Será necessário obter a série de Fourier da função $f:[0,2\pi)\to\mathbb{R}$. 

```python
from numpy import *
import matplotlib.pylab as plt
import matplotlib.colors
from pylab import cm
from scipy import integrate
from scipy.integrate import quad 
import numpy as np
```

Nesta parte, é feita a introdução da função $f$ (chamada de <font color="red">border</font> e são calculados os coeficientes da expansão de $f(\theta)$ em série de Fourier até ordem $N$ (por defeito, $N=20$). 

```python
border_input=input('Enter the function f (use Python 3 notation) as funcion of x ')
if border_input=="":
	print('Non valid equation')
	sys.exit()
    
border=lambda x:eval(border_input)

N=input('Enter the number of coefficients [20] ')
if N=="":
    N=20
else:
    N=int(N)

    

coscoef=[(1/(2*np.pi))*quad(border,-np.pi,np.pi)[0]]
sincoef=[0]

for i in range(1,N):
	coscoef+=[(1/np.pi)*quad(lambda t:border(t)*np.cos(i*t),-np.pi,np.pi)[0]]
	sincoef+=[(1/np.pi)*quad(lambda t:border(t)*np.sin(i*t),-np.pi,np.pi)[0]]
```

O próximo passo é definir as funções $\phi_{\mathrm{dentro}}$ e $\phi_{\mathrm{fora}}$, <font color="red">phidentro</font> e <font color="red">phifora</font>, respetivamente, representando a solução do problema no domínio $x^2+y^2=r^2\le 1$ e $x^2+y^2=r^2\ge 1$, respetivamente.

```python

def phidentro(theta,r):
#     return (sum([param[i]*r**i*np.cos(i*theta) for i in range(N)]))
     return (sum([r**i*(coscoef[i]*np.cos(i*theta)+sincoef[i]*np.sin(i*theta)) for i in range(0,N)]))

def phifora(theta,r):
     return (sum([r**(-i)*(coscoef[i]*np.cos(i*theta)+sincoef[i]*np.sin(i*theta)) for i in range(0,N)]))


def phi(theta,r):
	if r<=1:
		return(phidentro(theta,r))
	else:
		return(phifora(theta,r))
```

Finalmente, fazemos o gráfico da solução para valores de $r$ entre $0$ e $2$.

```python

#-- Generate Data -----------------------------------------
#azimuths = np.radians(np.linspace(0, 360, 180))
azimuths = np.radians(np.linspace(-180, 180, 180))
zeniths = np.arange(0, 2, 0.1)

r, theta = np.meshgrid(zeniths, azimuths)
values = [[phi(thetap,rp) for rp in zeniths] for thetap in azimuths]

#-- Plot... ------------------------------------------------
fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
p=ax.contourf(theta, r, values,50,cmap=cm.RdBu)
cb = fig.colorbar(p)

ax.plot(azimuths,[1 for t in azimuths],linewidth=3,color='black')

plt.tight_layout()
#plt.show()
```
